```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


data = {
    'date': ['1/1/2012', '1/2/2012', '1/3/2012', '1/4/2012', '1/5/2012'],
    'precipitation': [0.0, 10.9, 0.8, 20.3, 1.3],
    'temp_max': [12.8, 10.6, 11.7, 12.2, 8.9],
    'temp_min': [5.0, 2.8, 7.2, 5.6, 2.8],
    'wind': [4.7, 4.5, 2.3, 4.7, 6.1],
    'weather': ['drizzle', 'rain', 'rain', 'rain', 'rain']
}

df = pd.DataFrame(data)


df['date'] = pd.to_datetime(df['date'])

# Buat pivot table untuk heatmap (misalnya: temperatur maksimum per hari)
grouped_pivot = df.pivot_table(values='temp_max', index=df['date'].dt.month, columns=df['date'].dt.day, aggfunc=np.mean)

# Plot heatmap menggunakan pcolor
plt.figure(figsize=(6, 4))
plt.pcolor(grouped_pivot, cmap='RdBu', edgecolors='k')
plt.colorbar(label='Temperatur Maksimum (°C)')
plt.xlabel("Hari dalam Bulan")
plt.ylabel("Bulan")
plt.title("Heatmap Temperatur Maksimum")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

#DataFrame
data = {
    'date': ['1/1/2012', '1/2/2012', '1/3/2012', '1/4/2012', '1/5/2012'],
    'precipitation': [0.0, 10.9, 0.8, 20.3, 1.3],
    'temp_max': [12.8, 10.6, 11.7, 12.2, 8.9],
    'temp_min': [5.0, 2.8, 7.2, 5.6, 2.8],
    'wind': [4.7, 4.5, 2.3, 4.7, 6.1],
    'weather': ['drizzle', 'rain', 'rain', 'rain', 'rain']
}

df = pd.DataFrame(data)

# Buat scatter plot dengan regresi antara kecepatan angin dan temperatur maksimum
sns.regplot(x="wind", y="temp_max", data=df)


plt.ylim(0,)


correlation = df[["wind", "temp_max"]].corr()
print("Korelasi antara kecepatan angin dan temperatur maksimum:")
print(correlation)

# Tampilkan plot
plt.xlabel("Kecepatan Angin (m/s)")
plt.ylabel("Temperatur Maksimum (°C)")
plt.title("Regresi Linier: Kecepatan Angin vs Temperatur Maksimum")
plt.show()
```
